```python
#w4a7
from math import sqrt
def Primenumber(n):
    if n <= 1:
        return False
    for i in range (2,int(sqrt(n))+1):
        if n % i == 0 :
            return False
    return True
n = int(input())
for i in range (n , 1, -1):
    if n%i ==0 and Primenumber(i):
        print(i)
        break

```

```python
#w4a8
def reverse(n):
    t = 0
    while n>0:
        t = t * 10 + (n%10)
        n //= 10 
    return t 
n=int(input())
d = 0
while True:
    if n == reverse(n):
        break 
    n = n+ reverse(n)
    d += 1
print(d, n)

```

```python
#w4a9
from math import sqrt
def check (n):
    f = [0] * 10
    while n>0:
        if f[n%10]==0:
            f[n%10]=1 
        else:
            return False 
        n//=10
    return True
n = int(input())
for i in range(1, int(sqrt(n))):
    if check(i*i):
        print(i*i,end=" ")
if n == 0:
    print("no number")
```

```python
#w4a13
from math import sqrt
def DivisorSum(n):
    Sum = 0
    for i in range (1, int(sqrt(n)+1)):
        if n%i==0:
            Sum += i
            if i != n/i and i != 1:
                Sum += n/i
    return Sum
a, b= map(int,input().split())
if a == DivisorSum(b) and b == DivisorSum(a):
    print("True")
else:
    print("False")
```
