```python
#bai 4
h= int(input("Nhap chieu cao cua tam giac: "))
canh= h*2-1
for i in range(1,canh+1):
    if i==h:
        print("*",end="")
    else:
        print(" ",end="")
l=canh//2
r=canh//2 + 2
print('\n')
for i in range(1,h):
    if i == h-1:
        print("*"*canh)
    else:

        for i in range(1,canh+1):
            if i==l or i==r:
                print("*",end="")
            else:
                print(" ",end="")
        l-=1
        r+=1
    print('\n')
```

```python
#bai 20
import math
def check(n):
    if n<=1:
        return False 
    for i in range(2,int(math.sqrt(n))+1):
        if n%i==0:
            return False
    return True

n= int(input("Nhap so n: "))
count= 0
tmp= n
while tmp>0:
    x= tmp%10
    if check(x)==True:
        count+= 1
    tmp//= 10
print(f"{n} co {count} chu so nguyen to")
```
