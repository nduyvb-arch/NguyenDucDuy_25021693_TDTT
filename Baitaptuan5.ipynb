```python
#w5a3
from math import sqrt
def Primenumber(n):
    if n <= 1:
        return False
    for i in range (2, int(sqrt(n)) + 1):
        if n % i == 0 :
            return False
    return True
n = int(input())
if (Primenumber(n)):
    print(n, "la so nguyen to")
else:
    print(n, "khong la so nguyen to")
```

```python
#w5a4
from math import sqrt
def DivisorSum(n):
    Sum = 0
    for i in range (1, int(sqrt(n)+1)):
        if n%i==0:
            Sum += i
            if i != n/i and i != 1:
                Sum += n/i
    return Sum
n = int(input())
if (n == DivisorSum(n)) :
    print(n, "la so hoan hao")
else:
    print(n, "khong la so hoan hao")
```

```python
#w5a5
def solve (lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return '-1'    
lst = list(map(int, input().split()))
k = int(input())
print(solve(lst, k))
```

```python
#w5a6
def factorial (n):
        if n == 0:
            return 1
        else:
            return n * factorial(n-1)
n = int(input())
print(factorial(n))
```

```python
#w5a7 
def caculation (num1, num2, c): 
    if c == '+':
        return f"{num1 + num2:.2f}"
    if c == '-':
        return f"{num1 - num2:.2f}"
    if c == '*':
        return f"{num1 * num2:.2f}"
    if c == '/':
        return f"{num1 / num2:.2f}"
num1, num2 = map(float, input().split())
c = input()
print(caculation(num1, num2, c))
```

```python
#w5a8
def check (a, b):
    s1 = bin(a) [2:]
    s2 = bin(b) [2:]
    d = abs(len (s1) - len(s2))
    for i in range(len(s1)):
         if s1[i] != s2[i]:
              d += 1
    return d

a, b = map(int, input().split())
print(check(a, b))
```

```python
#w5a9
def tongcs (n):
    Sum = 0
    while n>0:
        Sum += n%10
        n //= 10
    
    return Sum
n = int(input())
print(tongcs(n))

```

```python
#w5a10
def check (s1, s2):

    f = [""] * 1000
    used = [False] * 1000
    if len(s1) != len (s2):
        return False
    for i in range (len(s1)):
        if f[ord(s1[i])] == "":
            if used [ord(s2[i])] == True:
                return False
            f[ord(s1[i])] = s2[i]
            used [ord(s2[i])] = True 
        else:
            if f[ord(s1[i])] != s2[i]:
                return False 
    return True

s1, s2 = map(str, input().split())
if check (s1, s2):
    print("True")
else:
    print ("False")
```
